```python
import pandas as pd

df = pd.read_csv('/content/student_performance_dataset_1000_rows (4).csv')

print("Shape:", df.shape)


pd.set_option('display.max_rows', 1000)
pd.set_option('display.max_columns', None)

display(df)
```

```python
print("Total Rows =", len(df))
```

```python

import matplotlib.pyplot as plt

df['Grade'].value_counts().plot(kind='bar')
plt.title('Grade Distribution')
plt.show()
```

```python
import matplotlib.pyplot as plt

df['Pass_Fail'].value_counts().plot(kind='bar', color=['green','red'])
plt.title('Pass/Fail Distribution')
plt.xlabel('Result')
plt.ylabel('Count')
plt.show()
```

```python
subjects = ['Tamil', 'English', 'Mathematics', 'Science', 'Computer']

df[subjects].mean().plot(kind='bar')
plt.title('Average Marks by Subject')
plt.xlabel('Subjects')
plt.ylabel('Average Marks')
plt.show()
```
